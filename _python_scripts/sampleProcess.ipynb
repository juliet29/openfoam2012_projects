```python
import os, sys
import math 
from numpy import loadtxt
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def add_y_v_ux(caseName, timeName):
    """ plotting y v ux"""
    fileName = caseName + "/postProcessing/singleGraph/" + timeName + "/line_U.xy"
    y = []
    ux = []
    abc = loadtxt(fileName, skiprows=4)
    for i in abc:
        y.append(i[0])
        ux.append(i[1])
    plt.plot(ux, y, label =timeName)

def add_y_v_ux(caseName, timeName):
    """ plotting y v k"""
    fileName = caseName + "/postProcessing/singleGraph/" + timeName + "/line_U.xy"
    y = []
    ux = []
    abc = loadtxt(fileName, skiprows=4)
    for i in abc:
        y.append(i[0])
        ux.append(i[1])
    plt.plot(ux, y, label =timeName)

def sorted_case_dirs(caseName):
    a = [int(dirStr) for dirStr  in os.listdir(f"{caseName}/postProcessing/singleGraph/")]
    a.sort()
    d = [str(i) for i in a]
    return d
```
