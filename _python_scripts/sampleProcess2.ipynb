```julia
import os, sys
import math 
from numpy import loadtxt
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```julia

def plot_time(root, time, x_dist=[0,10,20]):
    # x_dist = [0, 10, 20]

    u_dfs = {} # disctionary of data frames at each dist 
    for curr_dist in x_dist:
        u_path = os.path.join(root, str(time),  f"x_{curr_dist}m_U.xy")
        u_dfs[curr_dist] = pd.read_csv(u_path, delimiter="\t", names=["z", "Ux", "Uy", "Uz"])

    t_dfs = {} # disctionary of data frames at each dist, for turbulence proips
    for curr_dist in x_dist:
        t_path = os.path.join(root, str(time),  f"x_{curr_dist}m_epsilon_k_nut.xy")
        t_dfs[curr_dist] = pd.read_csv(t_path, delimiter="\t", names=["z", "epsilon", "k", "nut"])

    # plotting 
    fig, ax = plt.subplots(1, 3, sharey=True)

    # U 
    for curr_dist in x_dist:
        ax[0].plot(u_dfs[curr_dist].Ux, u_dfs[curr_dist].z, label=f"{curr_dist} m ")
        # print(curr_dist)
        ax[1].plot(t_dfs[curr_dist].k, u_dfs[curr_dist].z, label=f"{curr_dist} m ")
        ax[2].plot(t_dfs[curr_dist].epsilon, u_dfs[curr_dist].z, label=f"{curr_dist} m ")
    ax[0].set_xlabel("U (m/s)")
    ax[0].set_ylabel("z (m)")
    ax[2].legend()

    # ax[1].plot(obj.tke, obj.z)
    ax[1].set_xlabel("TKE (m2/s2)")

    # ax[2].plot(obj.epsilon, obj.z)
    ax[2].set_xlabel("Dissipation (m2/s3)")

    fig.suptitle(f"t = {time}")


    return u_dfs, t_dfs
    
```
